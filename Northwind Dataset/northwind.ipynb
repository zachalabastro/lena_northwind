```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ydata_profiling import ProfileReport
from ydata_profiling.config import Settings
```

```python
db = 'northwind.db'
conn = sqlite3.connect(db)
```

```python
all_data = pd.read_sql_query(
    "SELECT * FROM sqlite_master;",
    conn
)

# Show DataFields
all_data
```

```python
tables = pd.read_sql_query(
    "SELECT name FROM sqlite_master WHERE type='table';",
    conn
)

# Show Table Schemas
tables
```

```python
customers_df = pd.read_sql_query(
    "SELECT * FROM Customers;",
    conn
)

customers_df.head()
```

```python
orders_df = pd.read_sql_query(
    'SELECT * FROM Orders;',
    conn
)

orders_df.head()
```

```python
order_details_df = pd.read_sql_query(
    'SELECT * FROM "Order Details";',
    conn 
)

order_details_df.head()
```

```python
products_df = pd.read_sql_query(
    'SELECT * FROM Products;',
    conn
)

products_df.head()
```

```python
orders_enriched_df = orders_df.merge(order_details_df, on='OrderID', how='left'
    ).merge(products_df, on='ProductID', how='left'
    ).merge(customers_df, on='CustomerID', how='left')

orders_enriched_df.head()
```

```python
df_cleaned = (
    orders_enriched_df
    .drop(columns=[c for c in orders_enriched_df.columns if c.endswith('_y')])
    .rename(columns=lambda x: x.replace('_x', ''))
    # Important to keep data (if not ML).
    # .drop(columns=['Country', 'ShipCountry'], errors='ignore')
)

df_cleaned.head()
```

```python
df_cleaned['OrderRevenue'] = (df_cleaned['UnitPrice'] - (df_cleaned['UnitPrice'] * df_cleaned['Discount'])) * df_cleaned['Quantity']
df_cleaned = df_cleaned.reset_index(names='LineItem')
df_cleaned
```

```python
import base64
from ydata_profiling import ProfileReport

with open("logo.png", "rb") as f:
    image_data = base64.b64encode(f.read()).decode("utf-8")
    logo_base64 = f"data:image/png;base64,{image_data}"

with plt.style.context("fivethirtyeight"):
    profile = ProfileReport(
        df_cleaned, 
        title='LENA Northwind EDA',
        dataset={
            "description": "Personal EDA for the development of a data analytics framework under LENA.",
            "author": "Zach Alabastro",
            "name": "Northwind Dataset",
            "url": "https://github.com/jpwhite3/northwind-SQLite3?tab=readme-ov-file"
        },
        plot={"dpi": 200, "image_format": "png"},
        html={
            "navbar_show": True,
            "full_width": True,
            "style": {
                "theme": "flatly",
                "logo": logo_base64
            }
        }
    )

profile.to_file('enriched_northwind.html')
```
